```python
import torch
from PIL import Image
from captionGPT2 import captionGPT2
from decoder import GPTConfig
from encoder import VisionEncoderConfig
import tiktoken

# 1. Initialize model
gpt_config = GPTConfig(
    block_size=1024,
    vocab_size=50304,
    n_layer=12,
    n_head=12,
    n_embd=768,
    cross_attn_every=2
)
vision_config = VisionEncoderConfig(
    model_name="openai/clip-vit-base-patch32",
    projection_dim=768,
    freeze_encoder=True,
    dropout=0.1
)
model = captionGPT2(gpt_config, vision_config, freeze_gpt_base=True)
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

# 2. Load checkpoint
checkpoint = torch.load("caption_checkpoints/latest_checkpoint.pt", map_location="cpu")
model.load_state_dict(checkpoint['model'])
model.eval()

```

```python
# 3. Preprocess image
image = Image.open("single-women-happier-than-men-675ac891b545d.jpg").convert("RGB")
pixel_values = model.vision_encoder.image_processor(images=image, return_tensors="pt")["pixel_values"].to(device)

# 4. Prepare input_ids (start with <|endoftext|> or your BOS token)
tokenizer = tiktoken.get_encoding("gpt2")
input_ids = torch.tensor([[tokenizer._special_tokens['<|endoftext|>']]], dtype=torch.long).to(device)

# 5. Generation loop (greedy decoding example)
max_length = 5
for _ in range(max_length):
    with torch.no_grad():
        logits, _ = model(pixel_values, input_ids)
        next_token = logits[:, -1].argmax(-1, keepdim=True)
        input_ids = torch.cat([input_ids, next_token], dim=-1)
        if next_token.item() == tokenizer._special_tokens['<|endoftext|>']:
            break

# 6. Decode tokens to text
caption = tokenizer.decode(input_ids[0].tolist())
print("Generated caption:", caption)
```
